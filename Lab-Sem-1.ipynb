```python
#Reversing the bits of a number
def reverseBits(n) :
    rev = 0
    # traversing bits of 'n' from the right
    while (n > 0) :
        # bitwise left shift 'rev' by 1
        rev = rev << 1
        # if current bit is '1'
        if (n & 1 == 1) :
            rev = rev ^ 1
        # bitwise right shift 'n' by 1
        n = n >> 1
    # required number
    return rev
n = int(input())
print(reverseBits(n))
```

```python
# Fibonaci series nth term

n = int(input("Enter the numbe of terms: "))
n1, n2 = 0, 1
for i in range(n-1):
  nth = n1 + n2
  n2 = n1
  n1 = nth
print(n1)

# Fibonaci series upto nth term
for i in range(n):
  print(n1, end = ' ')
  nth = n1 + n2
  n2 = n1
  n1 = nth
```

```python
# Calculating total word count using map-reduce

from functools import reduce

def map_count(sentence):
  words = sentence.split()
  return len(words)

sentences = map(str, input().split('.'))
word_count = map(map_count, sentences)
total_word_count = reduce(lambda x,y: x + y, word_count)
print(total_word_count)
```

```python
# Give the path from source to destination. The user gives the intermediate path between source and destination as a nested list.

def find_path(L,s,d):
  if s == d:
    return [s]
  else:
    l = (s,d)
    if type(L[0]) != tuple:
      for i in L:
        if l in i:
          nl = [s,i[1],d]
          break
      return nl
    else:
      nl = [s,L[1],d]
      return nl

L = eval(input("Provide input list mentioning various sources and destinations: "))
s = int(input("Provide the source: "))
d = int(input("Provide the destination: "))
print(find_path(L,s,d))
```

```python
# Take the dictionary as an input and return the value for a specified key. Handle the KeyError exception.

def value_dict(d,k):
  try:
    value = d[k]
    return value
  except KeyError:
    print(f"KeyError: Key '{k}' not found in the dictionary.")
    return None

n = eval(input("Enter the dictionary: "))
k = input("Enter the key: ")
print("Result", value_dict(n,k))
```

```python
# WAP  that reads a text file named "numbers.txt" containing integers separated by spaces. Calculate and print the sum of those integers.

try:
  file_name = input("Enter the file name: ")
  with open(file_name, 'r') as file:
    numbers = [int(num) for num in file.read().split()]

  sum = sum(numbers)
  print(f"The sum of the numbers is {sum}")

except FileNotFoundError:
  print(f"File '{file_name}' not found.")
except ValueError:
  print("Invalid data in the file. Please ensure that the file contains integers separated by spaces.")
```

```python
#Create a class Circle with one attribute radius and two functions to compute the area and circumference of the circle.
#Print the area and circumference by initiating an object of the class.

class Circle():
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    return (self.radius**2)*3.14

  def circumference(self):
    return 2*self.radius*3.14

c = Circle(int(input("Enter the radius: ")))
print(round(c.area(),2))
print(round(c.circumference(),2))
```
