```python
# All permutations of a given string

def get_permuations(input_str, current_permuation = ''):
  if len(input_str) == 0:
    print(current_permuation)
  else:
    for i in range(len(input_str)):
      next_char = input_str[i]
      remaining_char = input[:i] + input_str[i+1:]
      get_permutations = (remaining_char, current_permuation + next_char)

input_str = input()
get_permuations(input_str)
```

```python
# Missing number from range
l = [1,2,3,4,5,6,7]
l1 = []
n1 = int(input("Enter the lower limit: "))
n2 = int(input("Enter the upper limit: "))
for i in range(n1,n2+1):
  for j in range(len(l)):
    if l[j] != i:
      l1.append(i)
      break
print(l1)
```

```python
# Perfect number

n = int(input())
sum = 0
for i in range(1,n):
  if n % i == 0:
    sum += i
if sum == n:
  print("The given number is Perfect number")
else:
  print("The given number is not Perfect number")
```

```python
# Use reduce to find the highest sum in a list of lists

from functools import reduce

l = [[1,2,3],[4,5,6],[7,8,9]]
max = 0
for i in l:
  sum = reduce(lambda x,y: x+y, i)
  if sum > max:
    max = sum
print(max)
```

```python
#Mirror alphabets of a string from a given position
s = input()
n = int(input())
l = list(s)
lst = []
for i in range(0,n-1):
  lst.append(l[i])
for i in range(n-1,len(l)):
  for j in range(97,123):
    if ord(l[i]) == j:
      lst.append(chr(122-(j-97)))
print(''.join(lst))
```
